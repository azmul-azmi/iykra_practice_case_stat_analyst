## Question 1: These are the countries where the guests come from

**First, we filter the quests with comfirm the booking**

```python
df_yes = df[df['is_canceled']==0]
```

**So, these are the top 5 countries where the quests come from**

```python
df_yes['country_origin'].value_counts().head()
```

## Question 2: These are the price of a room per night

**We have information about average cost per night per quest in column 'adr', so we use this value to calculate the price per room per night.**

```python
df_yes1 = df_yes[['hotel_type','arrival_date_year','arrival_date_month','arrival_date_day_of_month','market_segment','assigned_room_type', 'adr','reservation_status','reservation_status_date']]
df_pos = df_yes1[df_yes1['adr']>0]
```

**We create the pivot table group by hotel type and room type and calculate the average of 'adr' to find the price of a room per night.**

```python
pd.pivot_table(df_pos, index = ['hotel_type', 'assigned_room_type'], values = 'adr', aggfunc = 'mean')
```

## Question 3: The price of a room per night vary over year.

**We create the pivot table group by year, hotel type and room type and calculate the average of 'adr' to find the price of a room per night vary over year.**

```python
pd.pivot_table(df_pos, index = ['arrival_date_year','hotel_type', 'assigned_room_type'], values = 'adr', aggfunc = 'mean')
```

## Question 4: The busiest month.

**We create the pivot table group by year, month and calculate the count of 'adr' and sort by ascending to find the busiest month.**

```python
pd.pivot_table(df_pos, index = ['arrival_date_year', 'arrival_date_month'], values = 'adr', aggfunc = 'count').sort_values(by='adr', ascending = False)
```

## Question 5: Stay duration in hotel.

**To find the stay duration in hotel, we have to find the difference of arrive date and leave date in hotel and fill in this value to new column (create new column).** 

```python
df_pos['arrive'] = df_pos['arrival_date_day_of_month'].astype(str)+'/'+df_pos['arrival_date_month'].astype(str)+'/'+df_pos['arrival_date_year'].astype(str)
df_pos['arrive_date'] = pd.to_datetime(df_pos['arrive'], format='%d/%B/%Y')
df_pos['leave_date'] = pd.to_datetime(df_pos['reservation_status_date'], format='%d/%m/%Y')
df_pos['dur_stay'] = df_pos['leave_date'] - df_pos['arrive_date']
```

```python
df_pos['dur_stay'].astype('timedelta64[D]').value_counts().head()
```

```python
df_pos['dur_stay'].astype('timedelta64[D]').mean()
```

**So, average duration stay is 3 days - 4 days.**

## Question 6: Bookings by market segment.

**We create the pivot table group by market segment and calculate the count of reservation status to find the bookings by market segment.**

```python
pd.pivot_table(df_yes1, index = 'market_segment', values = 'reservation_status', aggfunc = 'count').sort_values(by='reservation_status', ascending = False)
```

## Question 7: Count of bookings that canceled.

**We filter the column 'is canceled' with value =1.** 

```python
df_no = df[df['is_canceled']==1]
df_no['is_canceled'].count()
```

**So, there are 44224 bookings that canceled.**

## Question 8: The month of highest number of canceled.

**We create the pivot table group by year and month and calculate the count of cancel status to find the The month of highest number of canceled.**

```python
pd.pivot_table(df_no, index = ['arrival_date_year', 'arrival_date_month'], values = 'is_canceled', aggfunc = 'count').sort_values(by='is_canceled', ascending=False)
```
